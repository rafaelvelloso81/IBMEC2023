```python
from mip import Model, maximize, xsum, BINARY, OptimizationStatus
```

Problema:


Problema.PNG

Montando Modelo:

```python
peso = [4, 6, 5, 3, 1]
valor = [5, 4, 3, 2, 1]
termo_independente = 10
```

```python
I = range(len(peso))
m = Model("Mochila")
```

```python
x = [m.add_var(var_type=BINARY) for i in I]
```

```python
m.objective = maximize(xsum(peso[i] * x[i] for i in I))
```

```python
m += xsum(valor[i] * x[i] for i in I) <= termo_independente
```

```python
status = m.optimize(max_seconds=5)
```

Modelo Otimizado?

```python
status == OptimizationStatus.OPTIMAL
```

**Resposta para o Problema**

```python
itens_selecionados = ["x"+str(i+1) for i in I if x[i].x >= 0.99]
print("Itens selecionados: {}".format(itens_selecionados))
```
